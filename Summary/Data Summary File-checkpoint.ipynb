```python
# Include the library

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os  
```

Next I made sure I can access all the files in the folder as give below,

```python
for fn in os.listdir('.'):
    if os.path.isfile(fn):
        print (fn)
```

Next I get the dataframe called Data which essentially contains the names of the files, the dimension of each file- to be interpreted as number of observations to numberof features and the data type of the variables present in each file.

```python
dataframe=[]
```

```python
for fn in os.listdir('.'):
    if os.path.isfile(fn):
        benchmark_data = pd.read_csv(fn, sep='\t')
        dataframe.append({'Dataset Name': fn, 'Dataset Dimension': benchmark_data.shape, 'Dataset Variables Type':
        benchmark_data.dtypes, 'Number of features': len(benchmark_data.columns)})
```

```python
Data=pd.DataFrame(dataframe)    
Data
```

Export this file to be saved as a csv file to retain a copy

```python
Data.to_csv("E:\Perelman Research\sklearn-benchmark\DataSummary File.csv")
```

So as can be seen there are 162 files in total, but some of these are repeated. But the copies of each file are alphabetically arranged. So I compare the number of features in each of the consecutive file as an indicator for similar files (since the features would be the same). and create a list of those files. 

Note: Perhaps comparing the datatypes of the files having the same number of features, in addition, would further indicate the redundancy

```python
same=[]
for i in range(0, len(Data)-1): 
    if Data['Number of features'].loc[i] == Data['Number of features'].loc[i+1]:
        #if Data['Dataset Variables Type'].loc[i] == Data['Dataset Variables Type'].loc[i+1]:
        same.append(i)      
```

Below shows that file number 2 is same as 3, 3 is same as 4, 4 is same as 5, 10 is same as 11 (with respect to the number of features). Crosschecking the dataframe confirms this observation.

```python
same
```

Below are the redundant files that can be removed from the original dataframe

```python
Remove=Data.ix[same]
Remove
```

Below is thus then the dataframes that are not repeated. It has 133 files as compared to 162 which were in the folder

```python
Data[~Data.isin(Remove)].dropna()
```
